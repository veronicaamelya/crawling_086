```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import time
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
data = {
    'Product Name': [],
    'Disc': [],
    'Harga awal': [],
    'Harga akhir': []
}
```

```python
browser.get('https://colorbox.co.id/collections/sale')
time.sleep(30)
soup = BeautifulSoup(browser.page_source, "html.parser")
rows = soup.find_all('div', class_='product-item__inner_')
```

```python
for row in rows:
    nama_element = row.find(class_='titledekstop')
    nama_Produk = nama_element.text if nama_element is not None else "None"

    disc_element = row.find(class_='flag-sale-text')
    disc = disc_element.text if disc_element is not None else "None"

    before_disc_element = row.find(class_='t-subdued')
    before_disc = before_disc_element.text if before_disc_element is not None else "None"

    after_disc_element = row.find(class_='sale sale-collection')
    after_disc = after_disc_element.text if after_disc_element is not None else "None"

    harga_element = row.find(class_='product-item__price price 1')
    harga = harga_element.text if harga_element is not None else "None"

    if disc != "None":
        harga_awal = before_disc
        harga_akhir = after_disc
    else:
        harga_awal = harga
        harga_akhir = harga

    data['Product Name'].append(nama_Produk)
    data['Disc'].append(disc)
    data['Harga awal'].append(harga_awal)
    data['Harga akhir'].append(harga_akhir)
```

```python
df = pd.DataFrame(data)
```

```python
df.to_excel('scrapping.xlsx', index=False)
```
